### Bronze notebook for exchange rate

```python
import requests
from datetime import datetime
import json


url = 'https://open.er-api.com/v6/latest/USD'

response = requests.get(url)

if response.status_code == 200 :

    data = response.json()
    filtered_data = {
    "time_last_update_utc": data.get("time_last_update_utc"),
    "rates": data.get("rates")
    }

    today = datetime.now().strftime("%Y-%m-%d")
    
    # Specify the file name and location to save json data into lakehouse
    file_path = f'/lakehouse/default/Files/{today}_exchangerate.json'

    # Open the file in write mode and dump the data into json format
    with open(file_path, 'w') as file:
        # json.dump writes directly to a file.
        # indent 4 makes human readable format
        json.dump(filtered_data, file,indent=4)

    print(f"Data is saved successfully to {file_path}")

else:
    print("failed to fetch data. Status Code", response.status_code )


```
