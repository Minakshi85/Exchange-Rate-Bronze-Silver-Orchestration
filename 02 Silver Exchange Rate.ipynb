### Silver layer Exchange rate

```python
from pyspark.sql.functions import col

df = spark.read.option("multiline", "true").json("Files/*.json")
# df now is a Spark DataFrame containing JSON data from "Files/2026-05-29_exchangerate.json".

# Extracting the attributes of features required for analysis purpose
df = \
df.\
select(
    'time_last_update_utc', \
    col('rates').getItem('USD').alias('USD'),\
    col('rates').getItem('AUD').alias('AUD'),\
    col('rates').getItem('INR').alias('INR'),\
    col('rates').getItem('EUR').alias('EUR'),\
    col('rates').getItem('CAD').alias('CAD'),\
    col('rates').getItem('JPY').alias('JPY')
)

# save the data in delta table
df.write.mode('overwrite').saveAsTable('silver_currency_exchange')
```
